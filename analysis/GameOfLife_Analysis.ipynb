```python
peri_np = "peri_20k_500_1_non_prime.txt"
peri_p = "peri_20k_500_1_prime.txt"
non_peri_np = "nperi_20k_500_1.txt"
peri_diff_size = "peri_diff_sizes.txt"
peri_diff_size1 = "peri_diff_sizes_1.txt"
```

## Processor Number Effects on Different Conditions



### Periodic vs Non-periodic
Generally the runtime trend of periodic and non-periodic conditions are similar. The runtime showed a logarithm decrearse which has approximately a slope of -1. However, the overall periodic runtime is 0.2 of the magnitude than it of non-periodic. The specific runtime of game of life and communication are further showed in the other two graphs. The apparent discrepancy between the game of life runtime of two conditions is probably from the implementation of the part. 

Communication time for periodic was in a relatively small range from $10^{-1}$ to $10^1$, yet fluctuating. The communication time is not fully representative of every other processors as the graph here only used time from processor 0. The non-periodic condition, on the other hand, had a trend of initially increasing from a very low communication time at processor number of 1. It is actually not a valid communication, but the time to get over the communication loop in the Source.cpp. The proportion of communication time for non-periodic condition will increase first as for a low processor number, the percentage of cells commiting less than 8 communications is higher than a domain which has a large size. The percentage of cells commiting 8 communications can be written as: 
${\frac{(m-2)\times(n-2)}{m\times n}}$, where $m \times n$ is the size of the domain.

```python
plt.figure(figsize=(5, 3), dpi=100, facecolor='w', edgecolor='k')
ax = plt.subplot(111)
m = np.array([2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 9192])
p = ((m-2)*(m-2))/(m*m)
ax.plot(m, p, 'k-o')
ax.set_title('Fig-1 Percentage for cells communicating 8 times')
plt.savefig('Fig_1.png')
```

## Efficiency comparison

The efficiency analysis uses the following equation: 
$$ E = \frac{1}{P}\frac{T_{serial}}{T_{parallel}}$$
where P is the number of processors, $T_{serial}$ is the runtime of serial code and $T_{parallel}$ is the runtime for parallelised code. 

This shows firstly that all timings are consistently valid among its own comparison. Non prime number of processors for both periodic and non-periodic conditions are generally higher (in the range from 0.8 - 1.0). It's necessary to notice that the use of efficiency here may not be fully representative as the parallelised runtime is only showed for processor 0. It appears that sometimes due to the discrepancy of core performance, runtime varies significantly (up tp 40% in this analysis). Here theoretically, if not considering the communication time between processors i.e. the overall workload can be directly assigned to several processors, the runtime should be: $$\frac{T_{serial}}{P}$$ 
which can be assumed to be the optimised time when parallelised. Therefore the ratio of the optimised runtime and the actual runtime could represent the efficiency of the corresponding parallel strategy. In other words, the reason why actual parallel runtime is different from the optimised one is due to the additional communication time. However, as the communication time shown in the timings are basically 2 - 3 magnitudes smaller than the actual game of life runtime, the efficiency was expected to be high for the first place. This is true for both periodic and non-periodic cases. However, when using prime number of processors, the efficiency decreases considerably to 0.5, which means the additional communication time used for stripes is higher than it for optimised partitioning with non-prime number of processors. Another reason is that, like claimed before, the use of only one processor can be coarse to generate a reasonable plot for efficiency. Actually other processors could sometimes be 1 magnitude smaller or larger the processor 0, which means that the actuall percentage for communication times varies, where the difference is larger when the number of processors is larger. In this aspect, when the processor number increases there is higher possibility for a core taking 1 mganitude longer communication time than other cores, causing other cores to wait for it (as MPI_Waitall was used in inplementation). This means that the exact time consumed for communication and its percentage in the total runtime could only be higher.

```python
# efficiency
pnp_eff = 625.671 / (pnp_data['proc_num'] * pnp_data[' total'])
pp_eff = 625.671 / (pp_data['proc_num'] * pp_data[' total'])
npnp_eff = 912.098 / (npnp_data['proc_num'] * npnp_data[' total'])

plt.figure(figsize=[10,5], dpi=100, facecolor='w', edgecolor='k')

plt.plot(pnp_data['proc_num'], pnp_eff, 'g-o', label='Periodic Non-Prime')
plt.plot(pp_data['proc_num'], pp_eff, 'r-o', label='Periodic Prime')
plt.plot(npnp_data['proc_num'], npnp_eff, 'b-o', label='Non-Periodic Non-Prime')

plt.grid(True)
plt.title('Fig-4 Efficiency Comparison\nMatrix size: 20k x 20k; iteration: 500')
plt.xlabel('Number of processors')
plt.ylabel('Efficiency')
plt.legend()
plt.savefig('Fig_4.png')
```

## With Different Sizes

Graphs below show the time and efficiency trend with increasing matrix sizes. The legend for LHS graph showed the polyfit result for the hyperbolic increase of runtime against domain size. 

Efficiency curve is as expected. For small domains, the higher the processor density the lower the efficiency as unnecessary runtime would be wasted for communications while the calculation for each processor is trivial comparing with its capability of calculation. This explaines why at the beginning the efficiency dropped to near 0: at that point communication time was actually longer than game of life runtime. One strange point appeared when solving a 10k by 10k domain for periodic condition: the efficiency was over 1.0 which is theoretically impossible. However, the result is valid as it' the average of 5 times running on HPC. One possible reason for this is that as there are plenty of cores available on HPC, calculation capability could vary significantly among them and in this case, the possibility was actually high as the processors used here is only 10.

For communication and game of life runtime, the RHS graph shows increasing trends of 1.3 and 2 (respectively). As domain size increases, communication time increases with less rapidly than game of life runtime which explains why the overall efficiency approaches 1. This again proves that when communication is low in percentage in the total runtime, efficiency appears to be high.

```python
# efficiency
pds_eff = pds_data1['total'] / (10*pds_data[' total'])

o2fit = np.polyfit(pds_data[' size'], pds_data[' total'], 2)
fit_result = o2fit[0]*(pds_data[' size']**2) + o2fit[1] * pds_data[' size'] + o2fit[0]

plt.figure(figsize=[15,6], dpi=100, facecolor='w', edgecolor='k')

ax1 = plt.subplot(121)
ax1.plot(pds_data[' size'], pds_data[' total'], 'go', label='Periodic Total')
ax1.plot(pds_data[' size'], fit_result, 'r--', label='Order 2 fit: \n%.2f\n%.2f\n%.2f'%(fit_result[0], fit_result[1], fit_result[2]))
ax3 = ax1.twinx()
ax3.plot(pds_data[' size'], pds_eff, 'b-o', label='Speedup')
ax1.grid(True)
ax1.set_title('Fig-5 Time vs Matrix Size\nIteration: 1000, Num of processors: 10, Memory: 10GB')
ax1.set_xlabel('Matrix size')
ax1.set_ylabel('Time used (s)')
ax1.legend()

ax2 = plt.subplot(122)
ax2.loglog(pds_data[' size'], pds_data[' comm'], 'c-o', label='Periodic Comm')
ax2.loglog(pds_data[' size'], pds_data[' game'], 'k-o', label='Periodic Game')
ax2.set_title('Fig-5 Time vs Matrix Size\nIteration: 1000, Num of processors: 10, Memory: 10GB')
ax2.set_xlabel('Matrix size')
ax2.set_ylabel('Time used (s)')
ax2.legend()
annotation.slope_marker((6e3, 1e1), (2, 1), ax=ax2,
                        size_frac=0.25, pad_frac=0.05, text_kwargs = dict(fontsize = 14))
annotation.slope_marker((8e3, 3e-1), (1.3, 1), ax=ax2,
                        size_frac=0.25, pad_frac=0.05, text_kwargs = dict(fontsize = 14))

plt.savefig('Fig_5.png')
```
